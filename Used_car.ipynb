### Preprocessing the data

```python
def split_it(x):
  try:
    return re.findall('[0-9.]+', x)[0]
  except:
    return np.nan

df['Mileage'] = df['Mileage'].apply(lambda x: split_it(x))
# df['Mileage'].isnull().sum()
```

```python
def spliter(x):
  try:
    return x.split(' ')[0]
  except:
    return np.nan
df['Engine'] = df['Engine'].apply(lambda x: spliter(x))
df['Engine'].isnull().sum()
```

```python
df['Power'] = df['Power'].apply(lambda x: split_it(x))
```

### Filling the null values

```python
df[['Mileage','Engine','Power']] = df[['Mileage','Engine','Power']].astype('f')
```

```python
# df.info()
brands = list(df.Brand.unique())
```

```python
for i in brands:
    df.Engine[(df.Engine.isnull()) & (df['Brand']==i)]= df.Engine[(df.Engine.notnull()) & (df['Brand']==i)].mean()
    df.Power[(df.Power.isnull()) & (df['Brand']==i)]= df.Power[(df.Power.notnull()) & (df['Brand']==i)].mean()
    df.Mileage[(df.Mileage.isnull()) & (df['Brand']==i)]= df.Mileage[(df.Mileage.notnull()) & (df['Brand']==i)].mean()

df.Seats[df.Seats.isnull()]=int(df.Seats[df.Seats.notnull()].mean())
```
